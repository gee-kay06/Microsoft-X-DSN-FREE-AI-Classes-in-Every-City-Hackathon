```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import log_loss
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('trainhackathon.csv')
df_test=pd.read_csv('Testhackathon.csv')
df.head()
```

```python
num = ['MONTANT', 'FREQUENCE_RECH', 'REVENUE', 'ARPU_SEGMENT', 'FREQUENCE', 'DATA_VOLUME', 'REGULARITY',
                      'ON_NET', 'ORANGE', 'TIGO', 'ZONE1', 'ZONE2', 'FREQ_TOP_PACK', 'CHURN']
cat= ['REGION', 'TENURE', 'MRG', 'TOP_PACK']
```

```python
df.describe()
```

```python
X = df.drop(columns=[ 'CHURN'])
y = df['CHURN']
```

```python
data=pd.concat([df, df_test], sort= False)
data.head()
```

```python
data.shape
```

```python
miss=data.isnull().sum()
miss
```

```python
misspct= (miss/len(data))*100
missdata= pd.DataFrame({'Missing':miss, 'Percent': misspct})
missdata
```

```python
data[cat].nunique()
```

```python
data['MRG'].value_counts()
```

```python
# Fill missing values for numerical columns with the mean
data[num] = data[num].fillna(data[num].median())

# Fill missing values for categorical columns with the mode
data[cat] = data[cat].apply(lambda x: x.fillna(x.mode()[0]))
data.isna().sum()
```

```python
data=data.drop(['user_id','ZONE1','ZONE2','REVENUE','MRG'],axis=1)
```

```python
data.info()
```

```python
data=pd.get_dummies(data, columns=['REGION','TENURE','TOP_PACK'], drop_first=True)
```

```python
data=data.drop(columns=['CHURN'],axis=1 )
```

```python
df= data.iloc[:len(y),:]
df_testa=data.iloc[ len(y):,:]
```

```python
X_train, X_test, y_train, y_test=train_test_split(df,y, test_size=0.1, random_state=11111)
```

```python
# Define a simplified model pipeline
model = Pipeline(steps=[
    ('classifier', LogisticRegression(max_iter=1000000, random_state=11111
                                        ))
])
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_proba= model.predict_proba(X_test)[:, 1]
logloss=log_loss(y_test, y_pred_proba)
logloss
```

```python
secondtest=pd.read_csv('Testhackathon.csv')
```

```python
subvalue=model.predict_proba(df_test)[:, 1]
```

```python
final_sub= pd.DataFrame({'user_id': secondtest['user_id'], 'CHURN': subvalue})
final_sub['CHURN']=final_sub['CHURN'].round(2)
final_sub.head(5)
```

```python
# Cross-validation to evaluate the model
cv_scores = cross_val_score(model, df, y, cv=5, scoring='neg_log_loss')
mean_cv_score = -cv_scores.mean()
```
